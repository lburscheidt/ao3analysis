```python
#biggest fandom by number of words posted
df = df.astype({'words':float})
total = df['words'].sum()
print(total)
```

```python
newdf = df.astype({'words':float}).fillna(0)

newdf.head()
```

```python
df1 = newdf[['fandoms', 'words']].fillna(0)
df2 = df1.explode('fandoms')
group = df2.groupby(['fandoms'], as_index=False)
df2.head()
```

```python
summed = group.sum()
print(summed.sort_values(by=['words'], ascending=False))
```
